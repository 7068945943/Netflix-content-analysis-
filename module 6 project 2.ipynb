## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Fill missing values in important columns
data['director'].fillna('Unknown', inplace=True)
data['cast'].fillna('Unknown', inplace=True)
data['country'].fillna('Unknown', inplace=True)
data['date_added'].fillna('Unknown', inplace=True)
data['rating'].fillna('Unknown', inplace=True)
````

````python
# Convert date_added to datetime
data['date_added'] = pd.to_datetime(data['date_added'], errors='coerce')
````

````python
# Remove leading/trailing spaces from object columns
for col in data.select_dtypes('object').columns:
    data[col] = data[col].str.strip()
````

````python
# Separate numeric part and type (Minutes or Seasons)
data['duration_int'] = data['duration'].str.extract('(\d+)').astype(int)
data['duration_type'] = data['duration'].apply(lambda x: 'Season' if 'Season' in x else 'Minute')
````

````python
# Convert genres to lowercase
data['listed_in'] = data['listed_in'].str.lower()
````

````python
data.drop_duplicates(inplace=True)
````

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatment
num_col = data.select_dtypes(include=['int64', 'float64']).columns
for col in num_col:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1

    # Define lower and upper limits
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    # Cap the outliers (Winsorization)
    data[col] = np.where(data[col] < lower_limit, lower_limit,
                         np.where(data[col] > upper_limit, upper_limit, data[col]))

# Check if outliers are treated
data.describe()
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

    I used the IQR (Interquartile Range) method for detecting and treating outliers.
    Values below Q1 – 1.5×IQR and above Q3 + 1.5×IQR were treated as outliers.

    To handle them, I used Winsorization (Capping) — replacing extreme values with the nearest boundary (lower or upper limit).
    This method was chosen because it reduces the impact of extreme values without deleting any data, keeping the dataset balanced and accurate for analysis.

````python
# Rephrase Text
import re

def rephrase_text(text):
    text = re.sub(r'(.)\1{2,}', r'\1', text)     # Repeated letters हटाना (e.g., "sooo" → "so")
    text = re.sub(r'!+', '!', text)               # Multiple exclamation हटाना
    text = re.sub(r'\?+', '?', text)              # Multiple question marks हटाना
    text = text.replace("’", "'").replace("‘", "'")  # Smart quotes normalize करना
    return text.strip()

data['clean_description'] = data['clean_description'].apply(rephrase_text)
````

````python
# Visualizing evaluation Metric Score chart
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Calculate Silhouette Score
sil_score = silhouette_score(scaled_data, kmeans.labels_)
print("Silhouette Score:", sil_score)

# Elbow Method to find optimal clusters
inertia = []
k_range = range(2, 10)

for k in k_range:
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(scaled_data)
    inertia.append(model.inertia_)

# Plot Elbow Curve
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.show()
````

````python
# Try different values of eps and min_samples
for eps in [0.3, 0.5, 0.7, 1.0]:
    for min_samples in [3, 5, 7]:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(x_scaled)
        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        if n_clusters > 1:
            silhouette = silhouette_score(x_scaled, labels)
            print(f"eps={eps}, min_samples={min_samples} -> Clusters={n_clusters}, Silhouette={silhouette:.3f}")
````
